```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npy file
data = np.load('combine_arc_scan/channel_0_format_6_idx_90_cfm.npy')

# Check if the data is 2D
if data.ndim == 2:
    # Plot the 2D image in grayscale with colormap range from 0 to 10
    plt.imshow(data, cmap='gray', aspect='auto', vmin=0, vmax=10)
    plt.colorbar(label='Intensity')
    plt.title("2D Image in Grayscale (0 to 10)")
    plt.xlabel("Column Index")
    plt.ylabel("Row Index")
else:
    print("Data is not 2D, unable to plot as an image.")

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Loop through the indices from 80 to 89
for idx in range(70, 80):
    # Construct the filename for each CFM file
    filename = f'combine_arc_scan/channel_0_format_6_idx_{idx}_cfm.npy'
    
    # Load the .npy file
    try:
        data = np.load(filename)
        
        # Check if the data is 2D
        if data.ndim == 2:
            # Plot the 2D image in grayscale with colormap range from 0 to 10
            plt.figure(figsize=(8, 6))
            plt.imshow(data, cmap='gray', aspect='auto', vmin=0, vmax=10)
            plt.colorbar(label='Intensity')
            plt.title(f"CFM {idx} - 2D Image in Grayscale (0 to 10)")
            plt.xlabel("Column Index")
            plt.ylabel("Row Index")
            plt.show()
        else:
            print(f"Data in {filename} is not 2D, skipping.")
    except FileNotFoundError:
        print(f"File {filename} not found, skipping.")
```
